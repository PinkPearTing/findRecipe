```python
from gensim.models import Word2Vec
import pandas as pd
import barbar
import json
```

```python
#
FILENAME = 'dat/processed/kaggle_and_nature_train'
```

```python
with open(FILENAME) as f:
    content = f.readlines()
# you may also want to remove whitespace characters like `\n` at the end of each line
recipe_in_lines = [x.split() for x in content]

# replace the '_' between word with space
recipe_list_with_space = []
for line in recipe_in_lines:
    recipe_list_with_space.append([word.replace('_', ' ') for word in line])
print('length of dataset: '+str(len(recipe_in_lines)))

# get the longest recepe and use this length as window in the later 
recipe_length_list = [len(line) for line in recipe_list_with_space]
max_length = max(recipe_length_list)
print('max length of recipe: '+str(max_length))
```

```python
recipe_list_with_space
```

```python
## load cleaned recipe data
df = pd.read_csv('cleaned_recipe.csv')
recipe_list = []
for id_value in set(df['id'].values):
    recipe_list.append(list(df[df['id']==id_value]['ingredient'].values))
    
max_length = max([len(recipe) for recipe in recipe_list])
```

```python
# senteces should be a lists of list of token in a sentence e.g. [['I','want','to','eat','apples'],[]...]
model = Word2Vec(sentences=recipe_list, size=100, window=max_length, min_count=1, workers=4)
```

```python
model.save('ingredient2vec.model')
```

```python
model.most_similar('salt')
```

```python
model.similarity('salt','white sugar')
```

```python
FILENAME = 'dat/train 2.json'
```

```python
# load data
with open('/Users/xueting/PycharmProjects/ingredientReplace/t', encoding="utf-8") as f:
    data = json.load(f)
```

```python
#
MINIMALSUPPORT = 3
# find the distribution of the ingredients
ingredient_set =set()
recipe_list = []
invalid_id_list = []
for i in range(len(data)):
    recipe = data[i]
    if len(recipe['ingredients'])>=1:
        for ingre in recipe['ingredients']:
            recipe_list.append([recipe['id'], ingre])

    else:
        invalid_id_list.append(recipe['id'])
```

## build the index for recipe id and ingredient

```python
recipe_dict = {}
food_to_recipe_id ={}
ingrelists = []
for recipe in data:
    recipe_dict[recipe['id']] = recipe['ingredients']
    ingrelists.extend(recipe['ingredients'])
    for ingre in recipe['ingredients']:
        if ingre in food_to_recipe_id:
            food_to_recipe_id[ingre].append(recipe['id'])
        else:
            food_to_recipe_id[ingre] = [recipe['id']]

for ingre in food_to_recipe_id:
    food_to_recipe_id[ingre] = list(set(food_to_recipe_id[ingre]))

with open('food_to_recipe_id.json', 'w') as outfile:
    json.dump(food_to_recipe_id, outfile)
with open('recipe_id_to_ingredients.json', 'w') as outfile:
    json.dump(recipe_dict, outfile)
```

## retrain the ingredient2vec.model

```python
recipe_list = list(recipe_dict.values())
max_length = max([len(recipe) for recipe in recipe_list])
model = Word2Vec(sentences=recipe_list, size=100, window=max_length, min_count=1, workers=4)
```

```python
model.save('ingredient2vecV3.model')
```
